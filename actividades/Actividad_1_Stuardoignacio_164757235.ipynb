## 2. Define en python vectores y funciones
<p align="justify">
Debes generar vectores que representan intervalos de tiempo o de una variable.
Debes ingresar las funciones matemáticas y evaluarlas en los vectores generados.

$$
f(x) =  e^{-x^2}
$$

$$
f(z) =  ysin(x)
$$

<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones
def f1(x):
    return np.exp(-x**2)

def f2(y, x):
    return y * np.sin(x)

# Generar los vectores
vector_1 = np.linspace(0, 10, 100)
vector_2 = np.linspace(5, 50, 100)

# Evaluar las funciones en los vectores
resultado_f1 = f1(vector_1)
resultado_f2 = f2(vector_2, vector_1)

# Graficar los resultados
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(vector_1, resultado_f1, color='blue')
plt.title('f(x) = e^(-x^2)')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplot(1, 2, 2)
plt.plot(vector_1, resultado_f2, color='red')
plt.title('f(z) = y*sin(x)')
plt.xlabel('x')
plt.ylabel('f(z)')

plt.tight_layout()
plt.show()

```

## 3. Ruido
<p align="justify">
Crear una señal (senoidal, cosenoidal, etc) y demuestra la agregación de ruido.
    


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal sinusoidal
amplitud = 1.0
frecuencia = 5.0  # Hz
tiempo_inicio = 0
tiempo_fin = 2.0  # segundos
numero_puntos = 1000

tiempo = np.linspace(tiempo_inicio, tiempo_fin, numero_puntos)

# Generar la señal sinusoidal
señal_sin = amplitud * np.sin(2 * np.pi * frecuencia * tiempo)

# Generar ruido aleatorio
ruido = np.random.normal(loc=0, scale=0.1, size=numero_puntos)

# Agregar ruido a la señal
señal_con_ruido = señal_sin + ruido

# Graficar la señal original y la señal con ruido
plt.figure(figsize=(10, 5))
plt.plot(tiempo, señal_sin, label='Señal original', color='blue')
plt.plot(tiempo, señal_con_ruido, label='Señal con ruido', color='red', alpha=0.7)
plt.title('Señal sinusoidal con ruido')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

El ruido rosa tiene la particularidad de que las frecuencias bajas de este poseen mayor potencia en el espectro, mientras que a mayor su frecuencia menor será su potencia.

La principal diferencia que existe entre el ruido gaussiano y el rosa es la distribucion de la potencia en el dominio de la frecuancia. El ruido gaussiano sigue una distribución de probabilidad gaussiana en el dominio del tiempo y la energía del ruido rosa está distribuida de manera inversamente proporcional a la frecuencia.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(10250)

ruido_gaussiano = np.random.normal(loc=0, scale=0.1, size=numero_puntos)
magnitud_ruido = np.abs(ruido_gaussiano)


def generate_pink_noise(samples):
    b = [0.049922035, -0.095993537, 0.050612699, -0.004408786]
    a = [1, -2.494956002, 2.017265875, -0.522189400]

    pink_noise = np.random.randn(samples)
    pink_noise = np.convolve(pink_noise, b)
    pink_noise = np.convolve(pink_noise, a, mode='valid')
    return pink_noise / 0.091420942

# generacion muestra ruido rosa
samples = 1000
pink_noise = generate_pink_noise(samples)

# FFT
spectrum = np.fft.fft(pink_noise)
freq = np.fft.fftfreq(samples)

# frecuencias positivas
positive_freq = freq[:samples // 2]
positive_spectrum = spectrum[:samples // 2]

# grafico ruido rosa
fig, ax = plt.subplots(2, 1, figsize=(10,6))
ax[0].plot(pink_noise)
ax[0].set_title("Ruido Rosa")
ax[0].set_xlabel("Tiempo")
ax[0].set_ylabel("Amplitud")

ax[1].plot(positive_freq, np.abs(positive_spectrum))
ax[1].set_xlabel("Frequencia")
ax[1].set_ylabel("Magnitud")

plt.figure(figsize=(10, 5))
plt.plot(tiempo, ruido_gaussiano, color='red')
plt.title('Ruido Gaussiano')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(tiempo, magnitud_ruido, color='red')
plt.title('Ruido Gaussiano')
plt.xlabel('Tiempo')
plt.ylabel('magnitud')
plt.grid(True)
plt.show()




```
